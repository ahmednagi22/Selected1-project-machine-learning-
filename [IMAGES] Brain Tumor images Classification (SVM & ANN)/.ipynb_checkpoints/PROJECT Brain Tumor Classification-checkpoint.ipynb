### Load Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### Prepare/collect data

```python
import os

path = os.listdir('brain_tumor/Training/')
classes = {'no_tumor':0, 'pituitary_tumor':1}
```

```python
import cv2
X = []
Y = []
for cls in classes:
    pth = 'brain_tumor/Training/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j, 0)
        img = cv2.resize(img, (200,200))
        X.append(img)
        Y.append(classes[cls])
```

```python
X = np.array(X)
Y = np.array(Y)

X_updated = X.reshape(len(X), -1)
```

```python
np.unique(Y)
```

```python
pd.Series(Y).value_counts()
```

```python
X.shape, X_updated.shape
```

### Prepare data

```python
X_updated = X.reshape(len(X), -1)
X_updated.shape
```

### Split Data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_updated, Y, random_state=0,
                                               test_size=.17)
```

```python
xtrain.shape, xtest.shape
```

### Feature Scaling

```python
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
```

### Train Model

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel="poly",degree=2,coef0=20,random_state=0)
sv.fit(xtrain, ytrain)
```

### Evaluation

```python
print("Training Score:", sv.score(xtrain, ytrain))
print("Testing Score:", sv.score(xtest, ytest))
ypred = sv.predict(xtest)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(ytest, ypred)
print("Confusion matrix: ")
print(cm)
print("Accuracy: ")
accuracy_score(ytest, ypred)
```

### Prediction

```python
pred = sv.predict(xtest)
```

```python
misclassified=np.where(ytest!=pred)
misclassified
```

```python
print("Total Misclassified Samples: ",len(misclassified[0]))
print(pred[36],ytest[36])
```

### TEST MODEL

```python
dec = {0:'No Tumor', 1:'Positive Tumor'}
```

```python
plt.figure(figsize=(12,8))
p = os.listdir('brain_tumor/Testing/')
c=1
for i in os.listdir('brain_tumor/Testing/no_tumor/')[:9]:
    plt.subplot(3,3,c)
    
    img = cv2.imread('brain_tumor/Testing/no_tumor/'+i,0)
    img1 = cv2.resize(img, (200,200))
    img1 = img1.reshape(1,-1)/255
    p = sv.predict(img1)
    plt.title(dec[p[0]])
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    c+=1
```
